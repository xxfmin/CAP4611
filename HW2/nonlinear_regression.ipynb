```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9.0, 6.0)
```

```python
# Preprocessing Input data
data = pd.read_csv('HW2_nonlinear_data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model

# Initialize coefficients
a = 0
b = 0
c = 0
d = 0

# Hyperparameters
L = 0.000001  # learning rate
epochs = 10000  # number of iterations
n = float(len(X))  # number of data points

# Gradient Descent for Cubic Function
for i in range(epochs):
    Y_pred = a * X**3 + b * X**2 + c * X + d
    
    # Gradients (partial derivatives of MSE)
    D_a = (-2/n) * sum(X**3 * (Y - Y_pred))
    D_b = (-2/n) * sum(X**2 * (Y - Y_pred))
    D_c = (-2/n) * sum(X * (Y - Y_pred))
    D_d = (-2/n) * sum(Y - Y_pred)
    
    # Update parameters
    a -= L * D_a
    b -= L * D_b
    c -= L * D_c
    d -= L * D_d

print(f"Y = {a:.4f}*X^3 + {b:.4f}*X^2 + {c:.4f}*X + {d:.4f}")
```

```python
# Making predictions
Y_pred = a * X**3 + b * X**2 + c * X + d

plt.scatter(X, Y, color='blue')
plt.scatter(X, Y_pred, color='red')  # predicted
plt.title("Non-Linear Regression")
plt.show()
```
